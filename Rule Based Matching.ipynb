**Matching can be done by matching tokens, matching phrases and matching entities. We can perform matching using the following methods.**

1. Create an object / instance of the Matcher Class.
2. Define a pattern (rule).
3. Add the pattern to the object we have created in the first step.
4. Pass the document to the object.

### Token Matching 

**We are trying to analyze the occurence of the text 'ChatGPT'.**

```python
# Importing the Matcher Class

from spacy.matcher import Matcher
```

```python
# Creating an instance of the Matcher Class

matcher_1 = Matcher(nlp.vocab)
```

```python
# Define a pattern or a rule

pattern_1 = [{'text' : 'ChatGPT'}]
```

**In spacy, a pattern is a list of dictionaries. Inside the dictionary, we have the attribute of the token (key) and the exact value of the token (value). In the above case we are trying to find the existence of the word 'ChatGPT' in the entire text.**

```python
# Add pattern to the object

matcher_1.add('Pattern1', [pattern_1])
```

**While adding the pattern to the object, we pass 2 arguments with the first one being a random one, while the second one being the pattern we have defined but passed inside a list.**

```python
# Pass the doc to the object

match_1 = matcher_1(doc)
```

**match_1 is the rule we are referring to using the pattern matcher_1 with the argument passed as the document.**

```python
# Printing the no of times 'ChatGPT' has been used in the document

print (len(match_1))
```

```python
# Printing the repetitions of 'ChatGPT' in the document.

for match_id, start, end in match_1:
    span = doc[start : end]
    print (span.text)
```

**A collection of token is called a span. match_id is the internal location of ChatGPT (shows the unique id by which the word ChatGPT is stored).**

### Phrase Matching

**We are trying to analyze the occurence of the phrase: 'ChatGPT is'**

```python
# Creating an object of Matcher Class and creating a pattern

matcher_2 = Matcher(nlp.vocab)
pattern_2 = [{'text' : 'ChatGPT'}, {'text' : 'is'}]
```

**Since we are analyzing for a phrase we will be requiring more than 1 dictionaries. The no of dictionaries is taken based on the number of matching you want to have.**

```python
matcher_2.add('Pattern2', [pattern_2])
```

```python
# Passing object to the matcher

match_2 = matcher_2(doc)
```

```python
# Checking the length

len(match_2)
```

**It states that 'ChatGPT is' is repeated 6 times.**

```python
for match_id, start, end in match_2:
    span = doc [start : end]
    print (span)
```

**LEMMA is nothing but the attribute of the token.**

```python
# Checking the count of repetitions of language model

for match_id, start, end in match_3:
    span = doc [start : end]
    print (span)
```

### Digit Matching

**We are trying to find the occurences of the digits preceded by the occurance of an alphabet.**

```python
matcher_4 = Matcher (nlp.vocab)
pattern_4 = [{'IS_ALPHA' : True}, {'IS_DIGIT' : True}]
matcher_4.add('Pattern4', [pattern_4])
match_4 = matcher_4(doc)
```

```python
# Checking the no of occurences 

print (len(match_4))
```

```python
# Checking the occurences of the alphabet and the digit

for match_id, start, end in match_4:
    span = doc[start : end]
    print (span)
```

### Occurence of launch, discovery, find etc.. (Synonyms)

```python
matcher_5 = Matcher(nlp.vocab)
pattern_5 = [{'LEMMA' : {'IN' : ['launch', 'discover', 'find', 'establish', 'invent','create', 'develop', 
                                 'innovate', 'form', 'initiate']}}]
matcher_5.add('Pattern5', [pattern_5])
match_5 = matcher_5(doc)
```

```python
# Counting the no of occurences

print (len(match_5))
```

```python
# Checking the occurences

for match_id, start, end in match_5:
    span = doc[start : end]
    print (span)
```
