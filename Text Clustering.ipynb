## Text Clustering

### Converting text into vectors

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf = TfidfVectorizer()
```

```python
features = tf.fit_transform(data)
```

```python
features
```

```python
print (features)
```

### Clustering using K-Means

```python
from sklearn.cluster import KMeans
```

```python
# Determining the no of clusters using kmeans

SSD = []
for k in range (1, 10):
    kmeans = KMeans(n_clusters = k, random_state = 10)
    kmeans.fit(features)
    SSD.append(kmeans.inertia_)

plt.plot(range (1, 10), SSD);
```

```python
# Determining the no of clusters using Silhouette score

from sklearn.metrics import silhouette_score

SS = []
for k in range (2, 10):
    kmeans = KMeans(n_clusters = k, random_state = 10)
    kmeans.fit(features)
    SS.append(silhouette_score(features, kmeans.predict(features)))

plt.plot(range(2, 10), SS);
```

**Cluster is always done to minimize the dataset and the analysis is easier with lesser no of clusters. Hence, we go with 5 nos of clusters.**

### Building the model with 5 Clusters

```python
kmeans = KMeans(n_clusters = 5, random_state = 10)
kmeans.fit(features)
```

```python
kmeans.labels_
```

### Updating the DataFrame with Cluster labels

```python
bbc['cluster'] = kmeans.labels_
bbc
```

**Since, we have a huge dimension, we go with tSNE as the Dimension Reduction technique.**

### Dimensionality reduction using TSNE

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components = 2, perplexity = 30, random_state = 10)
```

```python
features_tsne = tsne.fit_transform(features)
```

```python
features_tsne
```

```python
features_tsne.shape
```

### Visualization of the transformed Dataset

```python
plt.figure(figsize = (10, 8))
plt.scatter(features_tsne[:, 0], features_tsne[:, 1]);
```

```python
plt.figure(figsize = (10, 8))
plt.scatter(features_tsne[:, 0], features_tsne[:, 1], c = bbc['cluster'])
plt.legend(bbc['cluster']);
```

```python
fig, ax = plt.subplots(figsize = (10, 8))
sc = ax.scatter(features_tsne[:, 0], features_tsne[:, 1], c = bbc['cluster'], cmap = 'tab10')
ax.legend(*sc. legend_elements(), title = 'cluster');
```
